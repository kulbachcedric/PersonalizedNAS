# Add visualizations

```python
# Imports
import glob
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_paths = glob.glob("../results/*.csv")
file_paths = [Path(s) for s in file_paths]
```

```python
columns = []
result_data = pd.DataFrame()
for file_path in file_paths:
    splitted_name = file_path.name.split(sep='_')
    dataset_no = int(splitted_name[1])
    loss = splitted_name[-1].split(sep='.')[0]
    data = pd.read_csv(str(file_path))
    data['dataset'] = dataset_no
    data['loss'] = loss
    result_data = result_data.append(data)

```

```python
dataset_1_data = result_data[result_data['dataset']==2]
dataset_1_data = dataset_1_data[dataset_1_data['epoch']<=30]
sns.lineplot(data=dataset_1_data,hue='loss',x='epoch', y='precision')
plt.show()
```
